```
import pandas as pd 

uri='https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv'

data = pd.read_csv(uri)

print(data)
```

```
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score


x = data[['home', 'how_it_works','contact']]
y = data[['bought']]

SEED = 20

train_x, test_x, train_y, test_y = train_test_split(x,y, random_state = SEED,test_size = 0.25, stratify = y)

print("Trained with %d elements and tested with %d elements"%(len(train_x), len(test_x)))


model = LinearSVC()
model.fit(train_x, train_y)
predictions = model.predict(test_x)

accuracy = accuracy_score(test_y, predictions)*100

print("The accurary is %.2f%%"%accuracy)
```
